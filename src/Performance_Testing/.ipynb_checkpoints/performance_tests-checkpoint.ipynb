```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test.csv')
```

```python
print(df)
```

```python
#2D without alpha considered
mult1_data = df[df['MultType'] == 1]
print(mult1_data)
mult2_data = df[df['MultType'] == 2]
plt.plot(mult1_data['n'], mult1_data['AvgTime'], label='Mult1')
plt.plot(mult2_data['n'], mult2_data['AvgTime'], label='Mult2')
plt.xlabel('n')
plt.ylabel('Average Time')
plt.title("Comparing Times for Mult1 and Mult2")
plt.legend()
plt.show()
```

```python
#3D with alpha considered
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(mult1_data['n'], mult1_data['alpha'], mult1_data['AvgTime'], label='Mult1')
ax.scatter(mult2_data['n'], mult2_data['alpha'], mult2_data['AvgTime'], label='Mult2')
ax.set_xlabel('n')
ax.set_ylabel('alpha')
ax.set_zlabel('Average Time')
plt.title("Comparing Times for Mult1 and Mult2 with respect to n and alpha")
plt.legend()
plt.show()
```
