```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test.csv')
```

```python
print(df)
mult1_data = df[df['MultType'] == 1]
mult2_data = df[df['MultType'] == 2]
```

```python
#2D for total time without alpha considered
plt.scatter(mult1_data['n'], mult1_data['TotalTime'], label='Mult1')
plt.scatter(mult2_data['n'], mult2_data['TotalTime'], label='Mult2')
plt.xlabel('n')
plt.ylabel('Total Time (1/10 s)')
plt.title("Comparing Total Times for Mult1 and Mult2")
plt.legend()
plt.show()
```

```python
#2D for average time without alpha considered
plt.scatter(mult1_data['n'], mult1_data['AvgTime'], label='Mult1')
plt.scatter(mult2_data['n'], mult2_data['AvgTime'], label='Mult2')
plt.xlabel('n')
plt.ylabel('Average Time (ns)')
plt.title("Comparing Average Times for Mult1 and Mult2")
plt.legend()
plt.show()
```

```python
#3D with alpha considered
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(mult1_data['n'], mult1_data['alpha'], mult1_data['AvgTime'], label='Mult1')
ax.scatter(mult2_data['n'], mult2_data['alpha'], mult2_data['AvgTime'], label='Mult2')
ax.set_xlabel('n')
ax.set_ylabel('alpha')
ax.set_zlabel('Average Time (ns)')
plt.title("Comparing Average Times for Mult1 and Mult2 with respect to n and alpha")
plt.legend()
plt.show()
```
